```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("simulated_transaction_2024.csv")
```

```python
data['Date'] = pd.to_datetime(data['Date'], format='%d/%m/%Y')
```

```python
data["Date"] = pd.to_datetime(data['Date'], format="%m/%d/%Y")
data["Date"] = data['Date'].dt.strftime('%m/%d/%Y')
# Assuming data is your DataFrame containing dates
data["Date"] = pd.to_datetime(data["Date"], format="%m/%d/%Y")  # Convert to datetime

# Now perform datetime operations
data['Day_of_Week'] = data['Date'].dt.dayofweek
data['Month'] = data['Date'].dt.strftime('%B')
data['Week_of_Month'] = data['Date'].dt.day // 7 + 1

# You can convert 'Date' back to string format if needed
data["Date"] = data["Date"].dt.strftime('%m/%d/%Y')

# Display your DataFrame
print(data)
```

```python

plt.figure(figsize=(10, 6))
plt.scatter(data['Balance'], data['Amount'], s=10, c='blue', alpha=0.5)
plt.title('Scatter Plot of Amount vs balance')
plt.xlabel('Balance')
plt.ylabel('Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
from sklearn.ensemble import IsolationForest
selected_data = data[['Balance', 'Amount']]
selected_data.dropna(inplace=True)
# Create and fit the Isolation Forest model
model = IsolationForest(contamination=0.2)  # Adjust contamination based on expected percentage of outliers
model.fit(selected_data)
prediction=model.predict(selected_data)
# Predict anomaly scores
#anomaly_scores = model.decision_function(selected_data)

prediction
```

```python
outlier_indices = np.where(prediction < 0)[0]

# Plot the data
plt.figure(figsize=(10, 6))
plt.scatter(selected_data['Balance'], selected_data['Amount'], c='blue', label='Normal', s=10)
plt.scatter(selected_data.iloc[outlier_indices]['Balance'], selected_data.iloc[outlier_indices]['Amount'], c='red', label='Outlier', s=50)
plt.title('Isolation Forest Outlier Detection')
plt.xlabel('Balance')
plt.ylabel('Amount')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

from sklearn.preprocessing import StandardScaler
# Select columns of interest
selected_data = data[['Balance', 'Amount']]
selected_data.dropna(inplace=True)
scaler = StandardScaler()

# Fit and transform the data
scaled_data = scaler.fit_transform(selected_data)

# Convert the scaled data back to a DataFrame
scaled_df = pd.DataFrame(scaled_data, columns=['Scaled_Balance', 'Scaled_Amount'])


model = IsolationForest(contamination=0.2)  # Adjust contamination based on expected percentage of outliers
model.fit(scaled_df)

# Predict outliers
prediction = model.predict(scaled_df)

# Find indices of outliers
outlier_indices = np.where(prediction < 0)[0]

# Print outlier indices
print("Outlier indices:", outlier_indices)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(scaled_df['Scaled_Balance'], scaled_df['Scaled_Amount'], c='blue', label='Normal', s=10)
plt.scatter(scaled_df.iloc[outlier_indices]['Scaled_Balance'], scaled_df.iloc[outlier_indices]['Scaled_Amount'], c='red', label='Outlier', s=50)
plt.title('Isolation Forest Outlier Detection')
plt.xlabel('Scaled_Balance')
plt.ylabel('Scaled_Amount')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
anomaly_scores = model.decision_function(scaled_df)

# Plot the deviation from normal data
plt.figure(figsize=(10, 6))
plt.scatter(scaled_df['Scaled_Balance'], scaled_df['Scaled_Amount'], c=-anomaly_scores, cmap='viridis', s=10)
plt.colorbar(label='Anomaly Score')
plt.title('Isolation Forest Anomaly Detection')
plt.xlabel('Scaled_Balance')
plt.ylabel('Scaled_Amount')
plt.tight_layout()
plt.show()
```

```python
mean_balance = data['Balance'].mean()
std_balance = data['Balance'].std()

# Define thresholds
lower_threshold = mean_balance - 2 * std_balance
upper_threshold = mean_balance + 2 * std_balance

# Identify unusual balances
unusual_balances = data[(data['Balance'] < lower_threshold) | (data['Balance'] > upper_threshold)]

# Print unusual balances
print("Unusual Account Balances:")
print(unusual_balances)
```

```python
unusual_balances = data[(data['Balance'] < lower_threshold) | (data['Balance'] > upper_threshold)]
usual_balances = data[(data['Balance'] >= lower_threshold) & (data['Balance'] <= upper_threshold)]

# Plot the distribution of account balances
plt.figure(figsize=(10, 6))
plt.hist(usual_balances['Balance'], bins=50, color='blue', alpha=0.5, label='Usual Balances')
plt.hist(unusual_balances['Balance'], bins=50, color='red', alpha=0.5, label='Unusual Balances')
plt.axvline(x=lower_threshold, color='black', linestyle='--', label='Lower Threshold')
plt.axvline(x=upper_threshold, color='black', linestyle='--', label='Upper Threshold')
plt.xlabel('Account Balance')
plt.ylabel('Frequency')
plt.title('Distribution of Account Balances')
plt.legend()
plt.show()
```

```python
mean_balance = data['Balance'].mean()
std_balance = data['Balance'].std()

mean_amount = data['Amount'].mean()
std_amount = data['Amount'].std()

# Define thresholds for 'Balance' and 'Amount'
balance_lower_threshold = mean_balance - 2 * std_balance
balance_upper_threshold = mean_balance + 2 * std_balance

amount_lower_threshold = mean_amount - 2 * std_amount
amount_upper_threshold = mean_amount + 2 * std_amount

# Separate unusual and usual patterns
unusual_data = data[(data['Balance'] < balance_lower_threshold) | (data['Balance'] > balance_upper_threshold) |
                    (data['Amount'] < amount_lower_threshold) | (data['Amount'] > amount_upper_threshold)]
usual_data = data[~data.index.isin(unusual_data.index)]

# Plot the scatter plot with unusual and usual patterns
plt.figure(figsize=(10, 6))
plt.scatter(usual_data['Balance'], usual_data['Amount'], color='blue', alpha=0.5, label='Usual Patterns')
plt.scatter(unusual_data['Balance'], unusual_data['Amount'], color='red', alpha=0.5, label='Unusual Patterns')
plt.axvline(x=balance_lower_threshold, color='black', linestyle='--', label='Balance Lower Threshold')
plt.axvline(x=balance_upper_threshold, color='black', linestyle='--', label='Balance Upper Threshold')
plt.axhline(y=amount_lower_threshold, color='green', linestyle='--', label='Amount Lower Threshold')
plt.axhline(y=amount_upper_threshold, color='green', linestyle='--', label='Amount Upper Threshold')
plt.xlabel('Balance')
plt.ylabel('Amount')
plt.title('Unusual and Usual Patterns')
plt.legend()
plt.show()
```

```python
data['Date'] = pd.to_datetime(data['Date'])



# Calculate transaction frequency per account per day
transaction_frequency = data.groupby(['Account No', data['Date'].dt.date]).size().reset_index(name='Frequency')


# Calculate summary statistics for transaction frequency
mean_frequency = transaction_frequency['Frequency'].mean()
std_frequency = transaction_frequency['Frequency'].std()

# Define thresholds for unusual transaction frequency
lower_threshold = mean_frequency - 2 * std_frequency
upper_threshold = mean_frequency + 2 * std_frequency

# Flag unusual transaction frequencies
unusual_transactions = transaction_frequency[(transaction_frequency['Frequency'] < lower_threshold) | 
                                             (transaction_frequency['Frequency'] > upper_threshold)]

# Investigate unusual transaction frequencies
print("Unusual Transaction Frequencies:")
print(unusual_transactions)
```

```python
plt.figure(figsize=(10, 6))
plt.hist(transaction_frequency['Frequency'], bins=50, color='blue', alpha=0.5, label='Transaction Frequency')

# Plot thresholds for unusual transaction frequency
plt.axvline(x=lower_threshold, color='red', linestyle='--', label='Lower Threshold')
plt.axvline(x=upper_threshold, color='green', linestyle='--', label='Upper Threshold')

plt.xlabel('Transaction Frequency')
plt.ylabel('Frequency')
plt.title('Histogram of Transaction Frequency')
plt.legend()
plt.grid(True)
plt.show()
```
