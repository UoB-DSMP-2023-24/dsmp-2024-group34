```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("simulated_transaction_2024.csv")
```

```python
data['Date'] = pd.to_datetime(data['Date'])

data['time_diff'] = data['Date'].diff().dt.days

plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['time_diff'], marker='o', linestyle='-')
plt.xlabel('Date')
plt.ylabel('Time Interval (Days)')
plt.title('Time Intervals between Consecutive Dates')
plt.grid(True)
plt.show()

mean_time_diff = data['time_diff'].mean()
std_time_diff = data['time_diff'].std()

z_score_threshold = 3
outliers = data[(data['time_diff'] - mean_time_diff).abs() > z_score_threshold * std_time_diff]

print("Potential Outliers:")
print(outliers)
```

```python
data["Date"] = pd.to_datetime(data.Date,format="%d/%m/%Y")
data["Date"] = data['Date'].dt.strftime('%m/%d/%Y')
print(data['Date'])
```

```python
data['Day_of_Week'] = data["Date"].dt.dayofweek
data['Month'] = data['Date'].dt.strftime('%B')
data['Week_of_Month'] = data['Date'].dt.day // 7 + 1
data
```

```python
null_counts = data.isnull().sum()

print("Null values in each column:")
print(null_counts)
```

```python
total_rows = data.shape[0]
null_percentages = (null_counts / total_rows) * 100

print("Percentage of null values in each column:")
print(null_percentages)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

data['Date'] = pd.to_datetime(data['Date'])

data.set_index('Date', inplace=True)

decomposition = seasonal_decompose(data['Amount'], model='additive', period=30)
decomposition.plot()
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error

data['Date'] = pd.to_datetime(data['Date'])

data.set_index('Date', inplace=True)

model = ARIMA(data['Amount'], order=(5,1,0))
model_fit = model.fit()

forecast = model_fit.forecast(steps=30)

plt.plot(data.index, data['Amount'], label='Actual')
plt.plot(forecast.index, forecast, color='red', label='Forecast')
plt.xlabel('Date')
plt.ylabel('Amount')
plt.title('ARIMA Forecast')
plt.legend()
plt.show()


actual_values = data['Amount'][-30:]
mae = mean_absolute_error(actual_values, forecast)
print("Mean Absolute Error:", mae)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

X = data[['Timestamp']]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

kmeans = KMeans(n_clusters=3, random_state=0)
data['Cluster'] = kmeans.fit_predict(X_scaled)

plt.scatter(data['Timestamp'], [0] * len(data), c=data['Cluster'], cmap='viridis', alpha=0.5)
plt.xlabel('Timestamp ')
plt.title('Customer Segmentation')
plt.show()
```

```python
from sklearn.ensemble import IsolationForest


X = data[['Amount']]

isolation_forest = IsolationForest(contamination=0.05)
data['Fraud'] = isolation_forest.fit_predict(X)

plt.scatter(data.index, data['Amount'], c=data['Fraud'], cmap='coolwarm', alpha=0.5)
plt.xlabel('Date')
plt.ylabel('Amount')
plt.title('Fraud Detection')
plt.show()
```

```python
import networkx as nx

G = nx.DiGraph()

G.add_edges_from(zip(data['Account No'], data['Third Party Account No']))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=1000, edge_color='gray', arrowsize=10)
plt.title('Transaction Network')
plt.show()
```
